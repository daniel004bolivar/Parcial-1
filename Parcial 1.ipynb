```python
def scatter_regplot(data, strx, stry):
    sns.set(font_scale=1.4)
    fig, ax = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
    sns.scatterplot(data=data, x=strx, y=stry, ax=ax[0])
    sns.regplot(data=data, x=strx, y=stry, ax=ax[1]);
    fig.suptitle('Relación entre %s y VENTO, VELOCIDADE HORARIA (m/s)'%col)
```

```python
num_cols.remove('VENTO, VELOCIDADE HORARIA (m/s)')
for col in num_cols:
    scatter_regplot(data, col, 'VENTO, VELOCIDADE HORARIA (m/s)')
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import neighbors
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import r2_score
from sklearn.metrics import explained_variance_score
from sklearn.model_selection import cross_val_score
```

```python
X = data.drop('VENTO, VELOCIDADE HORARIA (m/s)', axis=1)
y = data['VENTO, VELOCIDADE HORARIA (m/s)']
```

```python
X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, random_state=0)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_trainval, y_trainval, random_state=1)
```

```python
print("Size of training set: {} size of validation set: {} size of test set: {}\n".
      format(X_train.shape[0], X_valid.shape[0], X_test.shape[0]))
```

```python
rmse_val = []
best_rmse = 1.0

for k in range(20):
    k = k + 1
    knn = neighbors.KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    if (rmse < best_rmse):
        best_rmse = rmse
        best_k = k
    rmse_val.append(rmse)
    print('RMSE value for k=', k, 'is:', rmse)

print(f"Best RMSE: {best_rmse}, Best k: {best_k}")
```

```python
knn = neighbors.KNeighborsRegressor(n_neighbors = 20)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
knn.score(X_test, y_test)
```

```python
r2_valid = r2_score(y_test, y_pred)
mae_valid = mean_absolute_error(y_test, y_pred)
rmse_valid = np.sqrt(mean_squared_error(y_test, y_pred))
print('Rsquare:', r2_valid)
print('RMSE Valid:',rmse_valid)
print('MAE_valid:', mae_valid)
```

```python
from sklearn.svm import SVC
X_trainval, X_test, y_trainval, y_test = train_test_split(X, y, random_state=0)
X_train, X_valid, y_train, y_valid = train_test_split(X_trainval, y_trainval, random_state=1)
print("Size of training set: {} size of validation set: {} size of test set: {}\n".
      format(X_train.shape[0], X_valid.shape[0], X_test.shape[0]))
```

```python
from sklearn.svm import SVR
gammas = [0.001, 0.01, 0.1, 1, 10, 100]
Cs = [0.001, 0.01, 0.1, 1, 10, 100]

best_rmse = 1.0
best_parameters = {}

for gamma in gammas:
    for C in Cs:
        svm = SVR(gamma=gamma, C=C)
        svm.fit(X_train, y_train)  # Ajuste del modelo SVR
        y_pred = svm.predict(X_valid)
        rmse = np.sqrt(mean_squared_error(y_valid, y_pred))  # RMSE para selección de parámetros
        if rmse < best_rmse:
            best_rmse = rmse
            best_parameters = {'C': C, 'gamma': gamma}  # Almacenamos el mejor RMSE y sus parámetros

print(f"Best RMSE: {best_rmse}")
print(f"Best Parameters: {best_parameters}")


        
```
